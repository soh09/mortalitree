```python
import re
import xml.etree.ElementTree as ET
from pathlib import Path

root = Path("weecology-NeonTreeEvaluation-f5d92da")
ann_dir = root / "annotations"
img_dir = root / "train"

# Training XMLs = no trailing "_NNN" plot index (i.e. they end in _image[_cropN]).
training_re = re.compile(r"_image(_crop[0-9]*)?\.xml$")
train_xmls = sorted(p for p in ann_dir.glob("*.xml") if training_re.search(p.name))
train_tifs = sorted(img_dir.glob("*.tif"))

xml_stems = {p.stem for p in train_xmls}
tif_stems = {p.stem for p in train_tifs}

matched = sorted(xml_stems & tif_stems)
xml_only = sorted(xml_stems - tif_stems)  # have label, no image
tif_only = sorted(tif_stems - xml_stems)  # have image, no label

print(f"Train XMLs:  {len(xml_stems)}")
print(f"Train TIFs:  {len(tif_stems)}")
print(f"Matched:     {len(matched)}\n")

print("=== MATCHED (xml + tif present) ===")
for s in matched:
    n = len(ET.parse(ann_dir / f"{s}.xml").getroot().findall("object"))
    print(f"  {s}  ({n} trees)")

print(f"\n=== XML ONLY — labeled but no image downloaded ({len(xml_only)}) ===")
for s in xml_only:
    n = len(ET.parse(ann_dir / f"{s}.xml").getroot().findall("object"))
    print(f"  {s}  ({n} trees)")

print(f"\n=== TIF ONLY — image but no label ({len(tif_only)}) ===")
for s in tif_only:
    print(f"  {s}")
```
